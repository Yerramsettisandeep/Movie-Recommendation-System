```python
def convert(obj):
    L = []
    for i in  ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def convert(obj):
    L = []
    counter3= 0
    for i in  ast.literal_eval(obj):
        if counter != 3: 
            L.append(i['name'])
            counter+=1
        else:
            break
        return L
```

```python
# Define the conversion function
def convert3(cast):
    # Your conversion logic here
    cast = eval(cast)[:3]
    l = [elem['name'] for elem in cast]
    return  l # For example, converting cast names to uppercase
# Apply the function to the 'cast' column
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L = []
    for i in  ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python

# removing spaces between words and splitting the string into a list of genres
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: ', '.join(x) if isinstance(x, list) else '')
movies['keywords'] = movies['keywords'].apply(lambda x: ', '.join(x) if isinstance(x, list) else '')
movies['cast'] = movies['cast'].apply(lambda x: ', '.join(x) if isinstance(x, list) else '')
movies['crew'] = movies['crew'].apply(lambda x: ', '.join(x) if isinstance(x, list) else '')

# Handle NaN in the 'overview' column
movies['overview'] = movies['overview'].fillna('')

# Concatenate columns to create 'tags'
movies['tags'] = (
    movies['overview'] + ' ' +
    movies['genres'] + ' ' +
    movies['keywords'] + ' ' +
    movies['cast'] + ' ' +
    movies['crew']
)
print(movies[['tags']])
```

VECTORIZATION

```python
#we r using  "bag of words" for removinf repeting words
#we r using stop_words for removing unwanted word like(in,are,a,..)
```

```python
ps = PorterStemmer()
```

```python
#creating short forms for loving=love,loved=love,...
#
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
#movie recommendation process
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        
        print(new_df.iloc[i[0]].title)
```
